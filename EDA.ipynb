```python
            #Exploratory Data Analysis(EDA)
#uncover patterns,detact anomalies and formulate hypotheses

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Read data from a csv file
rf=pd.read_csv('/content/Reinsuarance_data.csv')
print(rf.head())
```

```python
#Chcking the dimensions of the dataset
print("Shape of the dataset:",rf.shape)
```

```python
#Check Column names,data types and non-null counts
print(rf.info())
```

```python
#Summary statistics for numerical columns
print(rf.describe())
```

```python
#Check for missing values in thi case we don't have null values
print("Missing values:\n",rf.isnull().sum())
```

```python
Pol=rf['NO. OF POLICIES']
print(Pol)
```

```python
#Histogram or density plots for numerical variables
#ctrl+/ to multiple comment
Companies =rf['Company']

policies_count=rf["TYPE OF POLICY"].value_counts()
plt.figure(figsize=(20,15))
sns.histplot(rf,x="TYPE OF POLICY",multiple='stack',edgecolor='black')
plt.title("Distribution of Assured Amount by Company")
plt.xlabel("Type of policy")
plt.ylabel("Count per policy")
plt.grid()
```

```python
#Boxplots or violin plots to identify outliers and distribution
#Outliers;the process of identifying data points that have extreme values
#compared to the rest of the distribution.
plt.figure(figsize=(10,15))
sns.boxplot(x="Company",y='ASSUARED AMOUNT',data=rf)
plt.title("Boxplot of Numerical Column by Category")
plt.xlabel("Company")
plt.ylabel("ASSUARED AMOUNT")
plt.show()
```

```python
plt.figure(figsize=(10,15))
sns.scatterplot(x="Company",y='ASSUARED AMOUNT',data=rf,s=100)#s controls the marker size
plt.title("Boxplot of Numerical Column by Category")
plt.xlabel("Company")
plt.ylabel("ASSUARED AMOUNT")
plt.show()
```
